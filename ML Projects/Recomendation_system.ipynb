```python
import numpy as np
import pandas as pd
from surprise import Reader,Dataset,SVD
from surprise.model_selection import cross_validate
```

```python
ratings = pd.read_csv("/content/drive/MyDrive/DATASETS/combined_data_1.txt",header=None,names=['Cust_ID','Rating'],usecols=[0,1])
```

```python
ratings
```

```python
movie = pd.read_csv("/content/drive/MyDrive/DATASETS/movies.csv",encoding = "ISO-8859-1", usecols=[0,1,2])
```

```python
movie
```

```python
movie_id = None
movie_np = []

for i in ratings['Cust_ID']:
    if ':' in i:
        movie_id = int(i.replace(':', ''))
    movie_np.append(movie_id)
movie_np
```

```python
ratings["Movie_ID"] = movie_np
```

```python
ratings
```

```python
ratings["Rating"].isnull().sum()
```

```python
ratings["Movie_ID"].nunique()
```

```python
ratings.dropna(inplace=True)
```

```python
ratings
```

```python
movie_summary=ratings["Movie_ID"].value_counts()
```

```python
movie_summary
```

```python
rating_benchmark = round(movie_summary.quantile(0.6),0)
```

```python
rating_benchmark
```

```python
movie_id_rat = movie_summary[(movie_summary<=rating_benchmark)].index
```

```python
cust_sum = ratings['Cust_ID'].value_counts()
```

```python
cust_sum
```

```python
cust_benchmark = round(cust_sum.quantile(0.6),0)
```

```python
cust_benchmark
```

```python
movie_id_cust = cust_sum[(cust_sum<=rating_benchmark)].index
```

```python
ratings = ratings[(~ratings['Movie_ID'].isin(movie_id_cust)) & (~ratings['Cust_ID'].isin(movie_id_cust))]
```

```python
ratings
```

```python
reader= Reader()
```

```python
data = Dataset.load_from_df(ratings[['Cust_ID','Movie_ID','Rating']][:100000],reader)
data
```

```python
filter_ratings = ratings.groupby('Movie_ID')['Rating'].mean().sort_values(ascending=False)
```

```python
filter_ratings
```

```python
best_pop_gen_id = filter_ratings[filter_ratings>=4].index
```

```python
best_pop_gen_id
```

```python
popular_movies = movie[movie['movieId'].isin(best_pop_gen_id)]
```

```python
popular_movies
```

```python
best_rated_genres = popular_movies.merge(
    filter_ratings, left_on='movieId', right_index=True
).sort_values(by='Rating', ascending=False).reset_index(drop=True)
```

```python
best_rated_genres
```

```python
worst_gen_id = filter_ratings[filter_ratings<2.5].index
```

```python
bad_movies = movie[movie['movieId'].isin(worst_gen_id)]
```

```python
worst_rated_genres = bad_movies.merge(
    filter_ratings, left_on='movieId', right_index=True
).sort_values(by='Rating', ascending=False).reset_index(drop=True)
```

```python
worst_rated_genres
```

```python
model = SVD(random_state=42)
cross_validate(model,data,measures=['RMSE','MAE'],cv=3)
```

```python
def reccomend(id,n=5):
    user = movie.copy()
    user = user[(~user['movieId'].isin(movie_id_rat))]
    user['Estimate'] = user['movieId'].apply(lambda x: model.predict(id,x).est)
    return user.sort_values('Estimate',ascending=False)[['movieId', 'title', 'Estimate']].head(n)
```

```python
reccomend(1331154)
```

```python
def dont_reccomend(id,n=5):
    user = movie.copy()
    user = user[(~user['movieId'].isin(movie_id_rat))]
    user['Estimate'] = user['movieId'].apply(lambda x: model.predict(id,x).est)
    return user.sort_values('Estimate',ascending=False)[['movieId', 'title', 'Estimate']].tail(n)
```

```python
dont_reccomend(1331154)
```
