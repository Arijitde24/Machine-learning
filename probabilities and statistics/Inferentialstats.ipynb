```python
import numpy as np
from scipy import stats
```

#Z-test

##Two tail


```python
sample=50
sample_average=165
std=10
population_average=170
alpha=0.05

z=((sample_average-population_average)/(std/(sample**(1/2))))
p=2*(1-stats.norm.cdf(abs(z)))
print(f"z: {z}\n p: {p}")
```

```python
if p<alpha:
  print("reject null hypothesis, student height differs with average height")
elif p>alpha:
  print("fail to reject null hypothesis, student height does not differ with average height")
```

##One tail

```python
popaverage_marks=75
sampavg_marks=72
student_size=36
pop_std=8
alpha=0.05


z_score=(sampavg_marks-popaverage_marks)/(pop_std/(np.sqrt(student_size)))
p_value=stats.norm.cdf(abs(z_score))

print(f"z: {z_score}\n p: {p_value}")

if p_value<alpha:
  print("reject null hypothesis, average student marks < 75")
elif p_value>alpha:
  print("fail to reject null hypothesis, average student marks = 75")
```

#T-test

##Two Sample

```python
Before=[80,85,78,90,95,88,76,82]
After=[78,84,76,88,94,87,74,81]
alpha=0.05

t_score,p_value=stats.ttest_rel(Before,After)
print(f"t: {t_score}\n p: {p_value}")

if p_value<alpha:
  print("reject null hypothesis, significant effect on weight")
elif p_value>alpha:
  print("fail to reject null hypothesis, no significant effect on weight")
```

##One Sample

```python
avg_diameter=5
sample_size=10
diameter=[4.9,5.1,4.8,5.0,5.1,5.0,4.8]
alphaa=0.05

t_score,p_value=stats.ttest_1samp(diameter,avg_diameter)
print(f"t: {t_score}\n p: {p_value}")

if p_value<alphaa:
  print("reject null hypothesis, diameter differs from average")
elif p_value>alphaa:
  print("fail to reject null hypothesis, diameter does not differ from average")
```

#F-Test

```python
from scipy.stats import f
```

```python
out_1=[101,103,99,100,102,98]
out_2=[98,100,96,99,97,95]
f_value=np.var(out_1)/np.var(out_2)
alpha=0.05
dof1=5
dof2=5

p_value=1-f.cdf(f_value,dof1,dof2)
print(f"f: {f_value}\n p: {p_value}")

if p_value<alpha:
  print("reject null hypothesis, variances are significantly different")
elif p_value>alpha:
  print("fail to reject null hypothesis, variances are not significantly different")
```

#ANOVA

##One Way

```python
import pandas as pd
```

```python
data={
    'Sales' : [200,220,215,300,305,290,180,185,190,260,270,265],
    'Channel' : ['TV','TV','TV','Social Media','Social Media','Social Media',
                 'Radio','Radio','Radio','Radio','Radio','Radio']
}
df=pd.DataFrame(data)
df
```

```python
anova_stat,p_value=stats.f_oneway(df[df['Channel']=='TV']['Sales'],
                          df[df['Channel']=='Social Media']['Sales'],
                          df[df['Channel']=='Radio']['Sales'])

print(f"f: {anova_stat}\n p: {p_value}")

alpha=0.05

if p_value<alpha:
  print("reject null hypothesis, there is a significant difference in sales between marketing channels")
elif p_value>alpha:
  print("fail to reject null hypothesis, there is no significant difference in sales")
```

##Two Way

```python
data={
    'Sales' : [300,310,305,350,345,340,200,210,205,250,245,240],
    'Region' : ['North','North','North','South','South','South',
                'North','North','North','South','South','South'],
    'Channel' : ['TV','TV','TV','TV','TV','TV',
                 'Radio','Radio','Radio','Radio','Radio','Radio']
}

df=pd.DataFrame(data)
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model=ols('Sales~C(Region)+C(Channel)+C(Region):C(Channel)',data=df).fit() #Check if sales depends(~) on region or(+) channel or(+) region or interaction between channel and region from data df and to fit accordingly(.fit)
Anova_table=sm.stats.anova_lm(model,typ=2)
print(Anova_table)
```

```python
alpha=0.05

if Anova_table['PR(>F)']['C(Region)']<alpha:
  print('Reject null hypothesis, there is a significant interaction effect of region on sales')
elif Anova_table['PR(>F)']['C(Channel)']<alpha:
  print('Fail to reject null hypothesis, there is no significant interaction effect of channel and sales')
elif Anova_table['PR(>F)']['C(Region):C(Channel)']<alpha:
  print('Fail to reject null hypothesis, there is no significant interaction effect of region and channel on sales')
else:
  print('Fail to reject null hypothesis, there is no significant interaction effect of region and channel on sales')
```

#Chi-Square Test

##Test of Independence

```python
from scipy.stats import chi2_contingency

data={
    'Credit Card': [30,50],
    'Pay Pal' : [20,40],
    'Cash' : [50,30]
}

df=pd.DataFrame(data,index=['North','South'])
df
```

```python
chi2,p,dof,expected=chi2_contingency(df)
print(f"chi2: {chi2}\n p: {p}\n dof: {dof}\n expected: {expected}")
```

```python
alpha=0.05
if p<alpha:
  print('Reject Null Hypothesis, there is a significant relationship between payment methods and regions')
elif p>alpha:
  print('Fail to reject Null Hypothesis, there is no significant relationship between payment methods and regions')
```

##Goodness of fit

```python
observed=[25,45,30]
expected_prop=[0.3,0.5,0.2]
total_customer=sum(observed)
expected= [i * total_customer for i in expected_prop]


print(f"observed: {observed}\nexpected: {expected}\nexpected prop: {expected_prop}")
```

```python
chi,p=stats.chisquare(f_obs=observed,f_exp=expected)
print(f"chi: {chi}\n p: {p}")
```

```python
alpha=0.5
if p<alpha:
  print('Reject Null Hypothesis, the observed distribution does not match the expected distribution')
elif p>alpha:
  print('Fail to reject Null Hypothesis, the observed distribution matches the expected distribution')
```
