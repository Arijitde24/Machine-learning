##SAMPLING in INFERENTIAL STATISTICS##

```python
import pandas as pd
population=pd.Series(range(1,101))
sample=population.sample(10)
print(sample)
print("population mean ", population.mean())
print("sample mean ",sample.mean())
```

```python
#random sampling
random_sample=population.sample(10,random_state=42)#random_state gives fixed sample data according to the value provided to it
print("random sample \n",random_sample)
```

```python
#stratified sampling
from sklearn.model_selection import train_test_split
df=pd.DataFrame({
    'score':range(1,101),
    'Grade':['A']*30+['B']*40+['C']*30
})
#10% that goes into strat sample and 90% that goes into _ as we don't need it
strat_sample,_=train_test_split(df,test_size=0.9,stratify=df['Grade'])
print(strat_sample['Grade'].value_counts())
```

```python
#systematic sampling
data={
    'ID':range(1,101),
    'Value':range(100,200)
}

df=pd.DataFrame(data)
sample_size=10
population_size=len(df)
k=population_size//sample_size
start=3
systematic_sample=df.iloc[start::k]
print(systematic_sample)
```

##DESCRIPTIVE STATISTICS##

```python
import numpy as np
from statistics import mean,median,mode

data=[10,20,20,25,30,30,30,35,40,100]
print(len(data))
print(sum(data)/len(data))
print(mean(data))
if len(data)%2!=0:
  print(data[(len(data)+1)//2])
else:
  print((data[len(data)//2]+data[((len(data))//2)+1])//2)
print(median(data))
```

```python
if len(data)%2!=0:
  print(data[(len(data)+1)//2])
else:
  print((data[len(data)//2]+data[((len(data))//2)+1])//2)
print(median(data))
```

```python
# Step 1: Get unique elements
unique_elements = set(data)

# Step 2: Count occurrences and track max
max_count = 0
modes = []

for item in unique_elements:
    count = data.count(item)
    if count > max_count:
        max_count = count
        modes = [item]  # start fresh
    elif count == max_count:
        modes.append(item)  # add to mode list

print(modes)
```

```python
prob_class_A_given_passed=((data['Class']=='A') & (data['Passed']=='Yes')).mean()
print("Probability of passing given class A(p(passed|class=A))",prob_class_A_given_passed)
```

```python
Class_A=data[data['Class']=='A']
prob_A_passed=(Class_A['Passed']=='Yes').mean()
print("probability class A and Passed", prob_A_passed)
```

```python
#Normal
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

data=df['petal length (cm)']
x=np.linspace(data.min(),data.max(),100) # linspace equally distributes the data and takes input --> minimun value, maximum value, total no of items
mean=data.mean()
std=data.std()
y=norm.pdf(x,loc=mean,scale=std)#pdf means probability density function gives likeliness of getting a value in a continuous distribution, loc means my centre of grapf is mean and scale means spread is found using standard deviation
plt.hist(data,bins=20,density=True,alpha=0.6,color='skyblue')
plt.plot(x,y,'r',label='Normal Curve')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.title('Petal Length Distribution')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#.skew > 0 = positive skew
#,skew < 0 = negative skew
data=pd.Series([10,12,13,13,14,100])
print("Skkewness",data.skew())
```

```python
right_skewed=pd.Series([5,6,7,7,8,50])
print("Skewness",right_skewed.skew())
plt.hist(right_skewed, bins=10,color='orange',edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Right Skewed Distribution')
plt.show()
```

```python
left_skewed=pd.Series([1,20,21,22,23,24])
print("Left Skewness",left_skewed.skew())
plt.hist(left_skewed, bins=10,color='orange',edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Left Skewed Distribution')
plt.show()
```

```python
from scipy.stats import kurtosis
lepto=[10,10,10,10,100,10,10,10,10]
meso=np.random.normal(50,5,1000)
platy=[10,20,30,40,50,60,70,80,90]

print("Leptokurtosis",kurtosis(lepto))
print("Mesokurtosis",kurtosis(meso))
print("Platykurtosis",kurtosis(platy))
```

```python
meso=pd.Series(np.random.normal(50,5,1000))
print("Mesokurtosis",meso.kurt())

plt.hist(meso, bins=5,color='green',edgecolor='black')
plt.title('Mesokurtic Distribution')
plt.show()
```
